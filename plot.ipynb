```python
import matplotlib.pyplot as plt
import pandas as pd

def load_and_plot(ax, filename_top, filename_bottom, title, is_leftmost):
    # Load the data files into pandas DataFrames
    data_top = pd.read_csv(filename_top, sep=" ", header=None, names=["Carbon", "Scd"])
    data_bottom = pd.read_csv(filename_bottom, sep=" ", header=None, names=["Carbon", "Scd"])

    # Plot the data for top and bottom on the given axes object
    ax.plot(data_top["Carbon"], data_top["Scd"], marker='none', linewidth=3, label='No protein')
    ax.plot(data_bottom["Carbon"], data_bottom["Scd"], marker='none', linewidth=3, label='Protein-bound leaflet')

    # Add title, labels, and legend
    ax.set_title(title, fontsize=15)
    ax.set_xlabel('Carbon')
    if is_leftmost: 
        ax.set_ylabel('Scd', fontsize=20)
    ax.legend(fontsize=12)

# Replace 'file1.dat', 'file2.dat', 'file3.dat' with your .dat file names
files_top = ['POPC1_top_long.dat', 'POPC2_top_long.dat', 'DOPE1_top_long.dat']
files_bottom = ['POPC1_bottom_short.dat', 'POPC2_bottom.dat', 'DOPE1_bottom_short.dat']
titles = ['POPC1 Order Parameters', 'POPC2 Order Parameters', 'DOPE1 Order Parameters']

# Create a new figure with 3 subplots arranged in a horizontal stack
fig, axs = plt.subplots(1, 3, figsize=(18, 4))

# Adjust the horizontal spacing between the subplots
plt.subplots_adjust(wspace=0.2)

for i, (ax, file_top, file_bottom, title) in enumerate(zip(axs, files_top, files_bottom, titles)):
    load_and_plot(ax, file_top, file_bottom, title, i == 0)

# plt.savefig('op.png',dpi=300)
# Show the plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def load_and_plot(ax, filenames, labels, title):
    """
    Loads and plots the given files on the same axes.

    Parameters
    ----------
    ax : Axes
        The axes on which to plot the data.
    filenames : list of str
        The list of filenames to load and plot.
    labels : list of str
        The labels for each line in the plot.
    title : str
        The title for the plot.
    """
    # Load and plot each file
    for filename, label in zip(filenames, labels):
        # Load the data file into a pandas DataFrame
        data = pd.read_csv(filename, sep=" ", header=None, names=["Carbon", "Scd"])
        # Plot the data on the given axes object
        ax.plot(data["Carbon"], data["Scd"], marker='none', linewidth=2, label=label)

    # Add title, labels, and legend
    ax.set_title(title, fontsize=15)
    ax.set_xlabel('Carbon')
    ax.set_ylabel('Scd', fontsize=20)
    ax.legend(fontsize=12)

# Create a new figure
fig, ax = plt.subplots(figsize=(6, 4))

# The files to be plotted
filenames = ['POPC1_bottom.dat', 'POPC1_bottom_2nm.dat', 'POPC1_bottom_1nm.dat', 'POPC1_bottom_3nm_close.dat']
labels = ['>3 nm from protein', '>2 nm from protein', '>1 nm from protein', '<3 nm from protein']

# Load and plot the data
load_and_plot(ax, filenames, labels, 'POPC1 Order Parameters')

plt.savefig('op_close.png', dpi=300)
plt.show()
```
